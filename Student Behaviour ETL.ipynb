```python
#import dependencies
import pandas as pd
import numpy as np
import datetime
```

```python
file = "data/Student_Behaviour.csv"
df = pd.read_csv(file)
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={'Have you completed any certification courses, or are you currently enrolled in any?': 'Completed/Enrolled in Certification Course(s)?',
                        'college mark':'College Mark','hobbies':'Hobbies',
                        'possibility of choosing  their career based on their degree : ':'Probability of choosing career based on degree?',
                       'social medai & video games spending Time':'Time spent on Social Media/Video games',
                       'salary expectation':"Salary expectations", 'Are you doing a part-time job right now?':'Currently working part-time?', 
                        'daily studing time':'Daily studying time', 
                        'prefer to study in':'Time of day studying preference'})
```

```python
df.head()
```

```python
for i, row in df.iterrows():
    if df.loc[i,'Completed/Enrolled in Certification Course(s)?'] == "Yes":
        df.loc[i,'Completed/Enrolled in Certification Course(s)?'] = "Enrolled/Completed"
```

```python
df
```

```python
df.loc[:,'Salary expectations' ] = df['Salary expectations'].map('${:,}'.format)
```

```python
df.head()
```

```python
df.loc[:,'Financial Status'] = df['Financial Status'].str.capitalize()
```

```python
df.columns
```

```python
df = df.rename(columns = {"Stress Level ":"Stress Level"})
```

```python
df.loc[:,'Stress Level'] = df['Stress Level'].str.capitalize()
```

```python
df.head()
```

```python
df.to_csv('data/New_Student_Behaviour.csv')
```
